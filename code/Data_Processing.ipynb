### **Deleting Invalid Values**

```python
missing_labels = [
    "Residue",
    "Missing",
    "Missing / Not Asked",
    "Missing / Out of Universe",
    "Don't Know",
    "Refused"
]
df = df.replace(missing_labels, np.nan)
```

```python
## Given we already dropped the previous cols related to gang violence and it's high null count, we will drop the last one tied to it:
df = df.drop(columns = 'school_gang')
```

### **Imputing Mode to all values with NAs**

```python
##Imputting the mode for columns with missing values:
impute_cols = ['school_locale', 'school_type', 'school_type','other_drugs', 'illegal_drugs', 'painkillers_available', 'marijuana_available', 'alcohol_available',
                'saw_students_using_drugs', 'saw_students_using_drugs', 'access_to_gun', 'students_with_guns','grades', 'graffiti', 'knife_at_school', 'rumors', 'called_hate_speech', 'fear_at_school',
               'name_calling', 'social_exclusion', 'forced_actions', 'threats', 'physical_violence','property_crime', 'safe_at_school', 'school_level', 'race', 'current_grade'
               ]

for col in impute_cols:
  mode_val = df[col].mode()[0]
  df[col].fillna(mode_val, inplace=True)
```

### **Remapping as Integers**

```python
## Turning "Yes/no" columns back into binary 1/0
# List of all the columns you want to convert
binary_cols = [
    "name_calling", "rumors", "threats", "physical_violence",
    "forced_actions", "social_exclusion", "property_crime",
    "alcohol_available", "marijuana_available",
    "painkillers_available", "illegal_drugs", "other_drugs",
    "saw_students_using_drugs", "knife_at_school", "students_with_guns",
    "access_to_gun", "called_hate_speech", "graffiti"
]

binary_map = {'Yes': 1, 'No': 0}

for col in binary_cols:
    df[col] = df[col].map(binary_map)
```

```python
## We will do the same for "scaled" questions (Strongly Negative -> Strong Positive)

#Flipping the scale used by the codebook:
scale_map = {"Strongly Disagree": 1, "Disagree": 2,"Agree": 3,"Strongly Agree": 4}
scale_cols = ["crime_in_neighborhood", "safe_at_school"]

for col in scale_cols:
    df[col] = df[col].map(scale_map)
```

```python
## Same logic will be applied to fear_at_school:
fear_scale = {"Never": 1, "Almost never": 2,"Sometimes": 3, "Most of the time": 4}

df['fear_at_school'] = df['fear_at_school'].map(fear_scale)
```

### **Numpy Operations**

```python
## Checking for outliers:

#Only one numeric column, so we check ages:
df['age'].value_counts()
```

```python
## Calculating descriptive statistics for age:
age_len = df['age'].values

avg_age = np.mean(age_len)
med_age = np.median(age_len)

print('Average age', avg_age)
print('Median age', med_age)
```

Seems like our data could be slightly skewed towards younger children

```python
## Normalizing Ages:
min_age = np.min(age_len)
max_age = np.max(age_len)

df['norm_age'] = (df['age'] - min_age) / (max_age - min_age)
```
